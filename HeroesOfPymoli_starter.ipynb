```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame (CR: renamed it to include_df)
purchase_data_df = pd.read_csv(file_to_load)

#looking at head for visibility on dataset
purchase_data_df.head()
```

## Player Count

```python
#set variable to determine total player count using unique value counts
total_players =len(purchase_data_df["SN"].unique())

#print total player count
total_players

#unique player data frame
unique_players_df = pd.DataFrame({"Total Players": [total_players]})

unique_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# finding number of unique items by using len unique value counts
unique_items = len(purchase_data_df["Item ID"].unique())

#total purchases.. using unique values because each ID defined by number. will give total
purchase_total = len(purchase_data_df["Purchase ID"].unique())

#calculating total revenue with a sum
total_revenue = purchase_data_df["Price"].sum()

#average price
avg_price = round(total_revenue/purchase_total,2)


#summary data frame

summary_df = pd.DataFrame({"Number of Unique Items": [unique_items],
                           "Average Price": [avg_price],
                           "Number of Purchases": [purchase_total],
                           "Total Revenue": [total_revenue]})

#convert to floats and format
summary_df["Average Price"] = summary_df["Average Price"].astype(float).map("${:,.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].astype(float).map("${:,.2f}".format)

#display dataframe
summary_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#group data frame by SN
top_spenders = purchase_data_df.groupby("SN")

#set SN as index and do a count of purchases
sn_purchase_count = top_spenders["Price"].count()

#find total expenditure by SN
sn_total_expenditure = top_spenders["Price"].sum()

#calculate avg price by SN
sn_avg_price = sn_total_expenditure/sn_purchase_count

#create dataframe
sn_df = pd.DataFrame({"Purchase Count": sn_purchase_count,
                      "Average Purchase Price": sn_avg_price,
                      "Total Purchase Value": sn_total_expenditure})

#sort df by "Total Purchase Value" greatest to least
sn_df.sort_values(by=["Total Purchase Value"], ascending=False, inplace=True)

#convert to currency
sn_df["Average Purchase Price"] = sn_df["Average Purchase Price"].map("${:,.2f}".format)
sn_df["Total Purchase Value"] = sn_df["Total Purchase Value"].map("${:,.2f}".format)

#display using .head

sn_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Create a DataFrame using .loc to subset
item_df = purchase_data_df.loc[:, ['Item ID', 'Item Name', 'Price']]

# groupby the item DataFrame by "Item ID" and "Item Name" columns
item_group = item_df.groupby(["Item ID", "Item Name"])

# item purchase count
item_purchase_count = item_group['Price'].count()

#total item expenditure
item_total_expenditure = item_group['Price'].sum()

#individual item prices
item_price = item_total_expenditure/item_purchase_count

#create dataframe
item_summary = pd.DataFrame({"Purchase Count": item_purchase_count,
                             "Item Price": item_price,
                             "Total Purchase Value": item_total_expenditure})

#sort by "Purchase Count" greatest to least 
item_sorted = item_summary.sort_values(by=['Purchase Count'], ascending=False)

#convert to currency
item_sorted["Item Price"] = item_sorted["Item Price"].map("${:,.2f}".format)
item_sorted["Total Purchase Value"] = item_sorted["Total Purchase Value"].map("${:,.2f}".format)

#use .head to print
item_sorted.head()
```
