```python
# SQL Alchemy
from sqlalchemy import create_engine

# PyMySQL 
import pymysql
pymysql.install_as_MySQLdb()

import pandas as pd
```

```python
# Create Engine and Pass in MySQL Connection
engine = create_engine("mysql://root:<qwerty117>@localhost/restaurant_data")
```

```python
# Query All Records in the the Database
results = engine.execute("SELECT * FROM merged_table")

results = results.fetchall()

results
```

```python
results_df=pd.DataFrame(results)
```

```python
results_df = results_df.rename(columns={0: 'number_id',
                                  1: 'text_id',
                                  2: 'restaurant_name', 
                                  3: 'latitude',
                                  4: 'longitude',
                                  5: 'postal_code',
                                  6: 'pct_college_degree',
                                  7: 'pct_aa',
                                  8: 'pct_hispanic',
                                  9: 'hipanic_aa_pops',
                                  10: 'median_hh_income',
                                  11: 'total_population'})
```

```python
grouped_df = results_df.groupby(['restaurant_name']).mean()
```

```python
grouped_df.reset_index(inplace=True)
```

```python
grouped_df.to_sql('summary_table', con=engine, if_exists='replace',index_label='id')
```
